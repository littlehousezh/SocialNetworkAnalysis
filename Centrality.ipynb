```python
import networkx as nx
import matplotlib.pyplot as plt
import scipy as scipy
import numpy as np
import math
import networkx as nx
from scipy import stats, optimize, interpolate
import time
```

## Erdos-Renyi

```python
start = time.time()
ER_001 = nx.generators.random_graphs.gnp_random_graph(500,0.002)
nx.draw(ER_001,with_labels = False, node_size=10)
plt.title("ER Graph with n=500 and p=0.002")
plt.show()
end = time.time()
print(end - start)
```

```python
start = time.time()
ER_002 = nx.generators.random_graphs.gnp_random_graph(5000,0.0002)
nx.draw(ER_002,with_labels = False, node_size=10)
plt.title("ER Graph with n=5000 and p=0.0002")
plt.show()
end = time.time()
print(end - start)
```

## Barabasi-Albert

```python
start = time.time()
g1 = nx.generators.random_graphs.barabasi_albert_graph(500, 2)
nx.draw(g1,with_labels = False, node_size=30)
plt.title("BA Graph with n=500 and m=2")


plt.show()
end = time.time()
print(end - start)
```

```python
start = time.time()
g2 = nx.generators.random_graphs.barabasi_albert_graph(5000, 2)
nx.draw(g2,with_labels = False, node_size=30)
plt.title("BA Graph with n=5000 and m=2")


plt.show()
end = time.time()
print(end - start)
```

## Using the SciPy stats calculate both the correlation coefficients comparing the following centrality scores.

```python
scipy.stats.pearsonr(list(BA_closeness.values()), list(BA_betweenness.values()))
```

```python
scipy.stats.pearsonr(list(BA_closeness.values()), list(BA_eigenvector.values()))
```

```python
scipy.stats.pearsonr(list(BA_betweenness.values()),list(BA_eigenvector.values()))
```

```python
scipy.stats.spearmanr(list(BA_closeness.values()), list(BA_eigenvector.values()), axis=None, nan_policy='propagate')
```

```python
scipy.stats.spearmanr(list(BA_closeness.values()), list(BA_betweenness.values()), axis=None, nan_policy='propagate')
```
